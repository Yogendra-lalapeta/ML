Sales Dataset

```python
import pandas as pd
data = pd.read_csv('Datasets/salaries.csv')
data
```

```python
target = data['salary_more_then_100k']
target.head()
```

```python
data2 = data.drop('salary_more_then_100k',axis='columns')
data2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
company_label = LabelEncoder()
job_label = LabelEncoder()
degree_label = LabelEncoder()
```

```python
data2['company1'] = company_label.fit_transform(data2['company'])
data2['job1'] = job_label.fit_transform(data2['job'])
data2['degree1'] = degree_label.fit_transform(data2['degree'])
```

```python
data2.head()
```

```python
data2 = data2.drop(['company','job','degree'],axis="columns")
data2.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(data2,target)
```

```python
model.score(data2,target)
```

```python
val = model.predict([[2,2,0]])
val
```

Survival prediction in Titanic dataset

```python
ds = pd.read_csv("Datasets/titanic.csv")
ds.head(30)
```

```python
import math
med = math.floor(ds.Age.median())
med
```

```python
ds['Age'] = ds['Age'].fillna(med)
ds.head(6)
```

```python
le_Sex = LabelEncoder()
le_Fare = LabelEncoder()
le_Age = LabelEncoder()
```

```python
ds['SexLabel'] = le_Sex.fit_transform(ds['Sex'])
ds['FareLabel'] = le_Sex.fit_transform(ds['Fare'])
ds['AgeLabel'] = le_Sex.fit_transform(ds['Age'])
```

```python
ds = ds.drop(['Sex','Fare','Age'],axis='columns')
ds.head()
```

```python
target = ds['Survived']
```

```python

ds = ds.drop(['Survived'],axis='columns')
target.head()
```

```python
model2 = tree.DecisionTreeClassifier()
model2.fit(ds,target)
```

```python
model2.score(ds,target)
```
