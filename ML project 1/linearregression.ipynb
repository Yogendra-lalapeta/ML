```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_into_float(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return float(tokens[0])+float(tokens[1])/2
    try:
        return float(x)
    except:
        return None
        
        
```

```python
def remove_pps_outliers(ds1):
    ds_out = pd.DataFrame()
    for key,subds in ds1.groupby('location'):
        mean = np.mean(subds.price_per_sqft)
        stdie = np.std(subds.price_per_sqft)
        reduced_ds = subds[(subds.price_per_sqft>(mean-stdie))&(subds.price_per_sqft<=(mean+stdie))]
        ds_out = pd.concat([ds_out,reduced_ds],ignore_index='True')
    return ds_out
ds2 = remove_pps_outliers(ds1)
ds2.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
ds3 = remove_bhk_outliers(ds2)
# df8 = df7.copy()
ds3.shape
```

```python
ds3.head()
```

```python
ds4 = ds3[~(ds3.bath>ds3.bhk+2)]
ds4.shape
```

```python
ds4.location = ds4['location'].apply(lambda x:x.strip())
```

```python
dummies = pd.get_dummies(ds4.location,dtype=int)
dummies
```

```python
ds5 = pd.concat([ds4,dummies],axis='columns')
```

```python
ds6=ds5.drop(['other','location','size','price_per_sqft'],axis='columns')
ds6.head()
```

```python
X = ds6.drop('price',axis='columns')
y = ds6.price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=10)
regmodel = LinearRegression()
regmodel.fit(X_train,y_train)
regmodel.score(X_test,y_test)
```

```python
def find_price(location,sqft,bath,bhk):
    try:
        loc_ind = np.where(X.columns==location)[0][0]
    except:
        loc_ind = 0
    arr = np.zeros(len(X.columns))
    if loc_ind >=0:
        arr[loc_ind] = 1
    arr[0] = sqft
    arr[1] = bath
    arr[2] = bhk
    return regmodel.predict([arr])[0]
```
